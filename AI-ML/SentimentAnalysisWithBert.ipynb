```python
import tensorflow as tf
from transformers import TFBertForSequenceClassification, BertTokenizer
import numpy as np
```

```python
model = TFBertForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
tokenizer = BertTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
def analysis_emotions(text):
  print(f"Review: {text}")
  inputs = tokenizer.encode_plus(text, return_tensors='tf', max_length=512, truncation=True)

  output = model(inputs['input_ids'], attention_mask=inputs['attention_mask'])[0]
  prediction = tf.argmax(output, axis=1).numpy()[0]

  nature_map = {0: 'Very Negative', 1: 'Negative', 2: 'Neutral', 3: 'Positive', 4: 'Very Positive'}
  print(f'Emotions: {nature_map[prediction]}')
```

```python
product_reviews = [
    "Sound quality is too much good and value for money this product Highly recommend",
    "Bluetooth not Properly working",
    "Its impressive sound quality, sleek design and affordability make it a must-have for any home entertainment setup.",
    "Worst 👎 speakers don't buy this product, connecting is too worst",
    "Mivi sound bar Poor quality and this product is very very bad company",
    "Not good but ok ok",
    "The sound quality is awesome. I liked the balanced sound. Bass is powerful. Also connectivity is also good. It is more economical than other brands. Its value for money product. We can get the best sound quality in less amount. The clarity of voice is also good. Overall I recommend this product."
]
for i in product_reviews:
  analysis_emotions(i)
  print()
```
